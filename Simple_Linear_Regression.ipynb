`If the training_data is good -> after training -> if we get the actual points and the predictions are close means -> loss will be very low and accuracy will be high `

but if the training_data is not so good -> if actual points and prediction points are far away from each other we need to use Gradient Descent Optimizer

Gradient Descent will find the MSE with derivaties -> step size = c.f[MSE] + learning_rate [decide the speed to converge to global minumin point]

- New_slope -> old slope + s.s
- New_intercept -> old_intercept + ss


q28.gif

- `sample implementation `

```python
x_train = np.array([1,2,3,4,5])
y_train = np.array([5,7,9,11,13])

num_of_rows = len(x_train)

m_value = 0
c_value = 0

iterations = 10000
learning_rate = 0.04
plt.scatter(x_train,y_train,color='r')
for i in range(iterations):
  y_train_pred = m_value * x_train + c_value

  plt.plot(x_train,y_train_pred,color='g')
  cost_fun = (1/num_of_rows) * sum([i**2 for i in (y_train-y_train_pred)])
  m_der = -(2/num_of_rows)*sum(x_train * (y_train-y_train_pred))
  c_der = -(2/num_of_rows)*sum(y_train,y_train_pred)

  m_value = m_value - learning_rate * m_der
  c_value = c_value -learning_rate * c_der

  print(f'm : {m_value} and c : {c_value} iteration : {i}')

```
